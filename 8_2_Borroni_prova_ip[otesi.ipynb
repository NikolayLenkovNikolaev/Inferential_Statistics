## Test Uniformemente piu potenti

Il criterio della non distorsione e' ragionevole per un singolo test. Ma cosa succede quando vogliamo confrontare due test e sceliere il migliore?

Un idea potrebbe essere di scartare uno dei due se e' distorto ed in parte cio e' ragionevole , ma e'c hiaro che serve un criterio piu ampio

DEF:

Siano dati due test con fuznione di potenza $\beta_1(\theta)$ e $\beta_1(\theta)$ tali che

$sup_{\theta \in \Theta_0} \beta_1 (\theta) \leq \alpha$

$sup_{\theta \in \Theta_0} \beta_2 (\theta) \leq \alpha$

Il primo test e' uniformemente piu potente del secondo se:

$\beta_1(\theta \geq \beta_2(\theta)) $  all $\theta \in \Theta_1$

Per tanto, mettiamo al confronto solo test che abbiamo una dimensione inferiore e quelle desiderata . Trai due scegliamo quello che da la massima potenza(cioe la minore probabilta di Er-II) in qualunque punto di $\Theta_1$ (uniformemente)


Seguendo questa logica, si puo definire un test uniforme di ogni altro  e' necessario limitarsi a qulli con dimensione inferiore ad un livello $\alpha$ fissato e cercare quello con funzione di potenza uniformemente maggiore sui punti di $\Theta_1$

DEF:

Il test UNIFORMEMENTE PIUPOTENTE di livello $\alpha$ ha una funzione di potenza

$sup_{\theta \in \Theta_0}\beta^* (\theta)= \alpha$

e che

$\beta^* (\theta \geq \beta(\theta))$ for all $\theta \in \Theta_1$

Per ogni altro test con funzione di potenza che rispetti la restizione $sup_{\theta \in \Theta_0} \beta(\theta) \leq \alpha$

Purtroppo l acondizione contenuta nella definizione sopra e' molto difficile: molto spesso non esiste il test uniformememnte piu potente a meno che non si impongono ulteriori restrizioni ai test ammessi ad essere confrontati (e' un po quello che accade quando si cerca uno stimatore col minimo errore quadratico medio)
