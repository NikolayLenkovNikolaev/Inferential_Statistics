Def: uno stimatore per intervallo - da un parametro reale $\theta in \Theta$ e' definito da due funzioni campionarie:

$L(X_1...X_n)$ and $U(X_1...X_n)$

$L(x) \leq U(X)$ per ogni realizzazione campionaria $x=(X_1...X_n)$

La funzione di $\theta$

$P_{\theta} \in p[L(X); U(X)] = P_{\theta}[L(X) \leq \theta \leq U(X)]$

- detta PROBABILITA di COPERTURA

- l'estremo inferiore della prob di copertura e' detta Livello di Confidenza

In base alla definizione, una vlta che lo stimatore per intervallo e' stato costruitto, si ha una probabilita almento pari al livello di confidenza che il vero valore del parametro risulti nell intervallo.

Molto spesso avvioene che la prob.di copertura e' costante e non dipende da $\theta$. In ogni caso, quasi sempre si porte con un valore abbiettivo del livello di confidenza (indicato con $1-\alpha$) e; ltervallo si dice - INTERVALLO DI CONFIDENZA di $\theta$ al livello $1-\alpha$

## Stima Intervallare Bayesiana

Un problem delicato di interpretazione sorge quando occore interpretare una stima intervalare ottenuta secondo la definizione classica.

SUpponiamo ad esempio che uno stimatorre per intervallo sia tale per cui

$P_{\theta}[L(X) \leq \theta\leq U(X)]= 0.95$ all$\theta \in \Theta$ ---$*$

e che data una realizzazione campionaria x, si abbia

$L(x)= 10.8, U(x)= 12.2$

Saremmo tentati di concudere che

$P_{\theta}[10.8 \leq \theta \12,2]= 0.95$ all $\theta \in \Theta$ --- (**)

cio che IC[10.8; 12.2 ] contiene il parametro con una probabilita 95%

Tuttavio cio e' erato perche nella scrittura sopera none' nulla di aleatrio: benche non noto il parametro $\theta$ e' costante e dunque non comula di valopre a secondo del caro. Tale parametro si trova in [10.8; 12.2.= prob=1 ] o non si trova in [10.8, 12.2] Prob=0
 ma non puo trovarsi con una probabilita del 95%

 Il problemma interpretativo descritto non rifguarda invece gli interalli per $\theta$ costruiti seconda la logica della stat.bayesioana

 COme sappiamo in questo caso $\theta$ e' unavariabile aleatoria, quindi hanno senso affermaziooni come la second (**)

 $P(\theta \in A|x) = \int_A \Pi(\theta|x)d\theta=1-\alpha$
