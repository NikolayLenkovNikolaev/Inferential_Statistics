Talvolta e' possibili formulare delle congeture sul modello statistico (il che equivale a formulare sul parametro). I dati possono dunque essere utilizzati, piutosto che per stimare il parametro, semplicemente per avvalarne o confuteue le ipotesi poste:

DEF: un ipotesi statistica(Parametrica) e' una qualunque congettura riguardante un parametro

Esempi di ipotesi sono:
- $\theta=2$ o $\theta=\theta_0$
- $\theta 'leq 1$ o $\theta \leq \theta_0$


SOlitamente un'ipotesi che considera un solo punto dello spazio parametrico (es: $\theta=2$ se $\theta$ e' l'unico parametro) e che , dunque fissa univocamente il modello statistico e' detto SEMPLICE> IN caso COntrario - COMPOSTO ($\theta \ne 2$)

Una importante distinzione , in un problema di verifica di ipotesi e; quela tra ipotesi $H_0$ e $H_1$. Si possono considerare *paramentro formale*

- $H_0$  - cera fino a prova contraria


$H_0: \theta \in \Theta_0$ - ipotesi nulla

$H_1: \theta \in \Theta_1$

$\Theta_0; \Theta_1$ sono sottoinsiemi di $|theta$

Valori possibili di $\theta$ sono tali che:
- $\Theta_0 \cup \Theta_1 = \Theta$
- $\Theta_0 \cap \Theta_1 = \emptyset$

- $H_1$ congetura alternativa di $H_0$

- $H_0$ venga rifiutata. Questa prova e' una prova empirica , cioe e' basata sul dati campionari, secondo una regole della TEST


DEF: un TEST STATISTICO e' una regola decisionale che stabiisce quali realizzazioni campionari portino al rifiuto di $H_0$ (e dunque anche quali portino al rifiuto)


Un test statistico determina una sudivisione (in modo piu tecnico , uno aprtizione) dello spazio campionario

- regione di rifiuto - quali realizzazioni portono al rifiuto di $H_0$
- Regione di accetazione - quali realizzazioni non portano al rifiuto di $H_0$ - cio alla sua accettazione

Ovviamente, ogni decisione presa nei confronti di $H_0$ comporta una decisione nei confronto di $H_1$

Inoltre, poiche lo spazio campionario ha una struttura nota, per descrivere un test e' sufficiente descrivere la regione di rifiuto , dello spesso REGIONE CRITICA

Esempi:
- $R  \equiv [(x_1..x_n): \bar{x} \geq 2.58]$
- $R  \equiv [(x_1..x_n): |x_{(x)}- 6| < 0.22]$

Come si vede, spesso la regione critica e' basato sui valori assunti da una statistica, che pertanto e' dewtta STATISTICA TEST

Il valore con cui la stat.test e' confrontato viene spesso detto VALLORE CRITICO (possono esserci piu di un valore critico in untest)







### Test Basati sul rapporto di Verosimiglianza

COme ne; caso della stima puntuale da un lato ci sono dei metodi per trovare (propore) test, dell'altro metodi per confrontare test e sceglere quello migiore

Il metod piu utilizzato per trovere un test e' basato sulla verosimiglianza

DEF: Dato un problema di verifica di ipotesi in cui:
- $H_0: \theta \in \Theta_0$
- $H_1: \theta \in \Theta_1$

e $\theta = \Theta_0 \cup \Theta_1$ il test con la seguente regiuone di rifiuto

- $R \equiv [(x_1...x_n): \lambda(x) \leq c]$

dove 0<c<1

$\lambda(x) = \frac{ sup_{\theta \in \Theta_0} L(\theta|x) }{ sup_{\theta \in \Theta} L(\theta|x)}$

e' detto test del rapporto di verosimiglianza LRT- Likelihood ratio test

Logica: se l'ipotesi nulla contine valori del parametro con una verosimiglianza nettamente inferiore rispetto a quella che si puo ottenere in generale , significa che i dati nonsono a suporto di $H_0$, cio che questa iptesi aundrebbe rifiutata





Esempi:

consideriamo un campione $X_1...X_n$ IID da $f(x|\theta)= \frac{1}{\sqrt{2\pi}} exp[-\frac{1}{2} (x- \theta)^2]$

- $x \in R, \theta \in R$

- $H_0: \theta \in \theta_0$
- $H_1: \theta \in \theta_1$

Detterminiamo il test del rapporto di verosimiglianza

Abbiamo in primo luogo:

$L(\theta|x)= \prod_{i=1}^n  \frac{1}{\sqrt{2\pi}} exp[-\frac{1}{2} (x- \theta)^2]$

- constant - $ \frac{1}{\sqrt{2\pi}} $

La stima di massima verosimiglianza e' $\hat{\theta}= \bar{x}$

$sup_{\theta \in R} L(\theta|x) = L(\bar{x}|x)= exp[-\frac{1}{2} \sum_{i=1}^N (x_i - \bar{x})^2]$

Inoltre , sotto $H_0$ si ha un unico valore di $\theta: \theta_0$

Pertanto

$sup_{\theta = \theta_0} L(\theta|x)= L(\theta_0|x)= exp [- \frac{1}{2} \sum_{i=1}^n (x_i -\theta_0)]$

Il rapporto di verosimiglianza e'

$\lambda(x)= \frac{ exp [ -\frac{1}{2} \sum (x_i - \theta_0)^2] }{                   exp [ -\frac{1}{2} \sum (x_i - \bar{x} )^2] }$


$= exp [-\frac{1}{2}(\sum (x_i -\theta_0)^2 - \sum(x_i - \bar{x})^2)]$

Considerando ora un valore 0<c<1

$\lambda(x) \leq c -> exp[-\frac{1}{2} [\sum (x_i -\theta)^2 - \sum(x_i -\bar{x})^2]] \leq c$

$ exp[-\frac{1}{2} [\sum (x_i -\theta)^2 - \sum(x_i -\bar{x})^2]] \leq log\text{ c}$

$ exp[-\frac{1}{2} [n (x_i -\theta)^2]] \leq log\text{ c}$

$(\bar{x} -\theta_0)^2 \geq - \frac{2log\text{ c}}{n}$

$|\bar{x} -\theta_0| \geq \sqrt{- \frac{2log\text{ c}}{n}}$

Note:
- nei passafi si e' usato il fatto che, per ogni a

$\sum(x_i-a)^2 = \sum(x_i - \bar{x})^2 + n(\bar{x}-a)^2$

- essendo 0<c<1, $log\text{ c} < 0$ pertanto  $\sqrt{- \frac{2log\text{ c}}{n}}$
e' una costante reale che possiamo ridenominare come - k

Pertanto la regione critico del test del rapporto di verosimiglianza:
$R \equiv [(x_1..x_n): |\bar{x} -\theta_0| \geq k]$

## TEST BAYESIANI

Nell impostazione classica il parametro e' una costante che tuttavi nonconosciamo . Pertndo ad esempio , $H_0$ e' sicuramente vera o sicuramente falsa ma noi non sappiamo in quale contesto ci troviamo

$P(H_0 - vera) = P(\theta \in \Theta_0) =$ 0 o 1

Nell impoostazione bayesiano invece il parametro e' una variaile aleatoria e ha senso calcolare la probabilita che $H_0$ si avero con unnumero compresso tra 0 e 1

$P(H_0 - vera) = P(\theta \in \Theta_0) = \int_{\Theta_0} \pi(\theta)d \theta$

Questo probabilita esprime la nostra fiducia in $H_0$ ed e' basato sul solo giudizio. Tale fiducia puo conbinare per effetto delle informazioni fornite dai dati:

$P(H_0- vera|x) = P(\theta \in \Theta|x) \int_{\Theta_0} \pi (\theta) d\theta$

soi usera pertanto la distribuzione finale per calcolare questa probabilita. E' naturale decidere di rifiutare $h_0$ se il risultato e' una bassa probabilita , adalintura piu bassa di quella che si ottiene per $H_1$
Rifiuto $H_0$: $P(\theta \in\Theta_0 |x) < P(\theta \in \Theta_1|x)$

Poiche inoltre la somma delle due probabilta e' 2 si puo decidere solo ella scorta della prima:

Rifiuto $H_0$ se $P(\theta \in \Theta_0 |x) < 0.5$
- talvolta si e' piu conti e si pone una soglia piu bassa (es <0.1)






### Probabilita d'errore e' funzione di potenza

Una volta proposto un test, e' necessario valutare le sue potenzialita e' confronto con eventuali altri test per lo stesso problema.

Poiche il test comporta una decisone , un modo di valutare e' calcolare la probabilita che tale decisione sia errata- che dovrebbe essere basa se il testr e' buono

Se si accetta l'idea che $H_0$ sia da considerarrsi vera fino a prova contraria, l'errore di prima specie e' gerarchichamente piu grave di quello di seconda specie. Che usa il test, dunque verr controllare soprattuto la probailita di tale errore

$P(Er-I)= R(dex.rifiutare-H_0- qundo-vera) = P_{\theta} [(X_1...X_n)]$ e $|theta \in \Theta_0$

Analogamente si avra:

$P(Er-II) = P_{\theta}[(X-1...X_n) \notin  R]$ e $\theta \in \Theta_1$

E qunque importante capire che, per calcolare le due probabilita d'errore e' necessario fissare un valore di $\theta$(in $\Theta_0,\theta_1$) secondo del caso

Per tanto motivo, e' utile definire un'intera funzione di $\theta$ che serve e fornire queste probabilita

DEF: Data una regione critica R , si fef. FUNZIONE di POTENZA la funzione di $\theta$

$\beta(\theta)= P_{\theta} [(X_1...X_n) \in R]$

Pertanto potremo dire che

$
\begin{equation}
  \beta(\theta)=\begin{cases}
    P(Er-I), & \text{if $\theta \in \Theta_0$}\\
    1-P(Er-II) & \text{ if $\theta \in \Theta_1$}
  \end{cases}
\end{equation}
$

- $1-P(Er-II)$ coincide con la decisione corretta quando $H_1$ e' vera e' detto spesso POTENZA DEL TEST .

La funzione di potenza puo avere diverse forme a secondo dei casi . L'ideale sarebbe che fosse vicino a zero se $\theta \in \Theta_0$ e vicino a $1$ quando $\theta \in \Theta_1$
